```python
from aita.datasource.file import FileDataSource
```

```python
f = FileDataSource("../data/creditcard.csv", file_name="creditcard", file_format="csv")
```

```python
arrow_table = f.to_arrow()
```

```python
df = f.to_pandas()
df
```

```python
from aita.agent.pandas import PandasAgent

agent = PandasAgent("gpt-3.5-turbo").add_dataframe(dataframe=df)
```

```python
agent.chat("get me the first 10 rows of the dataframe in pandas")
```

```python
agent.chat(allow_run_tool=True)
```

```python
from aita.datasource.postgresql import PostgreSqlDataSource

pq = PostgreSqlDataSource(connection_url="postgresql://@localhost:5432/postgres")
```

```python
from aita.agent.pyarrow import PyArrowAgent

agent = PyArrowAgent("gpt-3.5-turbo").add_table(table=arrow_table).add_datasource(pq)
```

```python
agent.chat("ingest the dataframe into the database")
```
