```python
from ryoma_ai.datasource.file import FileDataSource
```

```python
f = FileDataSource("../data/creditcard.csv", file_name="creditcard", file_format="csv")
```

```python
arrow_table = f.to_arrow()
```

```python
df = f.to_pandas()
df
```

```python
from ryoma.agent.pandas import PandasAgent

agent = PandasAgent("gpt-3.5-turbo").add_dataframe(dataframe=df)
```

```python
agent.stream("get me the first 10 rows of the dataframe in pandas")
```

```python
agent.stream(allow_run_tool=True)
```

```python
from ryoma.datasource.postgres import PostgresDataSource

pq = PostgresDataSource(connection_url="postgresql://@localhost:5432/postgres")
```

```python
from ryoma.agent.pyarrow import PyArrowAgent

agent = PyArrowAgent("gpt-3.5-turbo").add_table(table=arrow_table).add_datasource(pq)
```

```python
agent.stream("ingest the dataframe into the database")
```
